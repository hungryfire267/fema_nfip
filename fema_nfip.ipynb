### Reducing the memory for our data

```python
"""
Creates two lists: 
- One float variable list that stores all the variables with a data type of float64
- One int variable list that stores all the variables with a datwa type of int64
"""
float_variables = []
int_variables = []
for feature in list(data.columns):
    data_type = data[feature].dtype
    if (data_type == "float64"): 
        float_variables.append(feature)
    elif (data_type == "int64"):
        int_variables.append(feature)

# Gets the minimum and maximum of each data type
min_float = min(data[float_variables].min())
max_float = max(data[float_variables].max())

min_int = min(data[int_variables].min())
max_int = max(data[int_variables].max())

print("Min float64 value:", min_float)
print("Max float64 value:", max_float)
print("Min int64 value:", min_int)
print("Max int64 value:", max_int)
```

```python
data[float_variables] = data[float_variables].astype(np.float32)
data[int_variables] = data[int_variables].astype(np.int16)

# Another look at the data type
data.info()
```

Great we can see that our overall dataset now is 300 megabytes less than the dataset previously.

```python
import re
def column_rename(variable):
    word_list = re.findall(r".[^A-Z]*", variable)
    word_list[0] = word_list[0].capitalize()
    feature = " ".join(word_list)
    return feature
```

#### Simplifying our datetime variables into several features

```python
import datetime as dt
```

```python
# List of Datetime variables
datetime_variables = [
    "As Of Date",
    "Date Of Loss",
    "Original Construction Date",
    "Original N B Date",
]
# Converting datetime variables from object to datetime
for variable in datetime_variables: 
    data[variable] = pd.to_datetime(data[variable], format="%Y-%m-%dT%H:%M:%S.%fZ")
    print(variable, ":", data[variable].dtype)
```

```python
day_variables = []
month_variables = []
year_variables = []
time_variables = []
for variable in datetime_variables: 
    # Day
    day_variable = variable.replace("Date", "Day")
    data[day_variable] = data[variable].dt.day
    day_variables.append(day_variable)
    # Month
    month_variable = variable.replace("Date", "Month")
    data[month_variable] = data[variable].dt.month
    month_variables.append(month_variable)
    # Year
    year_variable = variable.replace("Date", "Year")
    data[year_variable] = data[variable].dt.year
    year_variables.append(year_variable)
    # Time
    time_variable = variable.replace("Date", "Time")
    data[time_variable] = data[variable].dt.time
    time_variables.append(time_variable)
    # Delete the original variable
    del data[variable]
```

Now, let us plot another boxplot of our three target variables, but this time we remove all the outliers.

```python
def remove_outliers(feature, data): 
    mean = data[feature].mean()
    std = data[feature].std()
    lower_limit = mean - 1.96 * std
    upper_limit = mean + 1.96 * std
    data_feature = data[feature][(data[feature] >= lower_limit) & (data[feature] <= upper_limit)]
    og_counts = data[feature].notnull().sum()
    new_counts = data_feature.notnull().sum()
    no_outliers = og_counts - new_counts
    percentage_outliers = no_outliers/data.shape[0]
    og_missing = data[feature].isnull().sum()
    percentage_missings = og_missing/data.shape[0]
    return data_feature, no_outliers, percentage_outliers, lower_limit, upper_limit, percentage_missings
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
outlier_target_dict = {}
features, outliers, percentage_outliers, lower_limits, upper_limits, percentage_missings = [], [], [], [], [], []
for i, feature in enumerate(target_variables): 
    data_feature, no_outliers, percentage_outlier, lower_limit, upper_limit, percentage_missing = remove_outliers(feature, data)
    features.append(feature)
    outliers.append(no_outliers)
    percentage_outliers.append(round(percentage_outlier * 100, 3))
    lower_limits.append(lower_limit)
    upper_limits.append(upper_limit)
    percentage_missings.append(round(percentage_missing * 100, 3))
    sns.boxplot(y=data_feature, ax=axs[i])
    axs[i].set_title(feature, fontsize=13)
outlier_target_dict["Features"] = features
outlier_target_dict["No. Outliers"] = outliers
outlier_target_dict["Percentage of Outliers"] = percentage_outliers
outlier_target_dict["Acceptable Lower Limit"] = lower_limits
outlier_target_dict["Acceptable Upper Limit"] = upper_limits
outlier_target_dict["Percentage Missing"] = percentage_missings
fig.suptitle("Univariate boxplot of target features (outliers removed)", fontsize=16)
plt.tight_layout()
plt.show()
```

##### Transforming our data

Transforming our data to a more normal distribution will not only improve homogeneity to our data, but also lead to much more accurate predictions. If transformation of our target variables does not occur, then it is possible that our model cannot predict the right tail accurately (but can predict the variables that have a true value being closer to 0 much more accurately).

To find the best transformation of our data to a roughly normal distribution we test the following transformations: 
- Cubic Root
- Fifth Root
- Hepta Root

Note: We will not transform our `Account Paid on Increased Cost of Compliance Claim` as we do not have a detailed distribution of this variable because:
- We can see that there are a lot of values at $30, 000 for the `Amount Paid on Increased Cost of Compliance Claim`. This is due to there being a cap (of $30, 000) to "help cover the cost of mitigation measures that will reduce the flood risk". 
- Roughly $97\%$ of all observations do not have an observation for this target feature.

```python
fig, axs = plt.subplots(3, 2, figsize=(10, 15))
root_transforms = [1/3, 1/5, 1/7]
for i, transform in enumerate(root_transforms):
    for j, variable in enumerate(target_variables[:2]): 
        data_feature, _, _, _, _, _ = remove_outliers(variable, data)
        sns.histplot(x = data_feature ** transform, bins=20, ax=axs[i, j])
        axs[i, j].set_title(f"Transform of {variable} with {round(root_transforms[i], 3)}")
fig.suptitle("Effect of transforms on our target data")
plt.tight_layout()
plt.show()
```

Looking at our we can see that the transform of $0.2$ is the most optimal transform since it balances skewness and normality very well. The transform of $0.333$ left a postive skew on both of our transformed target data variables, while the transform of $0.143$ left a negative skew on both of our transformed target data variables.

```python
def plot_datetime_target(features):
    fig, axs = plt.subplots(3, 3, figsize=(24, 24))
    fig.suptitle("Our loss datetime features against our target variables")
    fig.supxlabel("Target Variables")
    fig.supylabel("Loss feature labels")
    for i, target in enumerate(target_variables):
        no_outlier_target, _, _, _, _, _ = remove_outliers(target, data)
        final_target = no_outlier_target ** (1/5)
        for j, feature in enumerate(features): 
            final_variable = data[feature][final_target.index]
            new_df = pd.concat([final_variable, final_target], axis=1)
            mean = new_df.groupby(by=feature).mean()
            std = new_df.groupby(by=feature).std()
            lower = (mean - 1.96 * std).reset_index()
            upper = (mean + 1.96 * std).reset_index()
            mean = mean.reset_index()
            sns.lineplot(
                data=lower, x=feature, y=target, ax=axs[i, j], label="95% Lower Limit", color="#CCB974"
            )
            sns.lineplot(
                data=mean, x=feature, y=target, ax=axs[i, j], label="Mean", color="#4C72B0", linestyle="dashed",
            )
            sns.lineplot(
                data=upper, x=feature, y=target, ax=axs[i, j], label="95% Upper Limit", color="#CCB974"
            )
            axs[i, j].fill_between(lower[feature], lower[target], upper[target], color="#FFF1BF", alpha=0.3)
            axs[i, j].set_title(f"{feature} vs {target}")
    plt.show()
```

#### Data Imputation

```python
features = list(new_data.columns[2:])
new_data.isna().sum()
```

```python
def datetime_imputations(new_data):
    original_construction = ["Original Construction Day", "Original Construction Month", "Original Construction Year"]
    for feature in original_construction:
        feature_mode = new_data[feature].mode().iloc[0]
        new_data[feature] = new_data[feature].fillna(feature_mode)
    original_nb = ['Original N B Day', 'Original N B Month', 'Original N B Year']
    for feature in original_nb:
        feature_mode = new_data[feature].mode().iloc[0]
        new_data[feature] = new_data[feature].fillna(feature_mode)
    return new_data
    
def boolean_imputations(new_data, boolean_variables): 
    for feature in boolean_variables: 
        feature_mode = new_data[feature].mode().iloc[0]
        new_data[feature] = new_data[feature].fillna(feature_mode)
    return new_data

def categorical_imputations(new_data):
    # Too Many Missing values for Basement Enclosure Crawlspace - Delete
    del new_data["Basement Enclosure Crawlspace"]
    # Too Many Missing values for Basement Enclosure Crawlspace - Delete
    del new_data["Community Rating System Discount"]
    obstruction_type_mode = new_data["Obstruction Type"].mode().iloc[0]
    new_data["Obstruction Type"] = new_data["Obstruction Type"].fillna(obstruction_type_mode)
    rate_method_mode = new_data["Rate Method"].mode().iloc[0]
    new_data["Rate Method"] = new_data["Rate Method"].fillna(rate_method_mode)
    #### FOR NOW - TO CHANGE
    flood_zone_mode = new_data["Flood Zone"].mode().iloc[0]
    new_data["Flood Zone"] = new_data["Flood Zone"].fillna(flood_zone_mode)
    no_floors_mode = new_data["Number Of Floors In The Insured Building"].mode().iloc[0]
    new_data["Number Of Floors In The Insured Building"] = new_data["Number Of Floors In The Insured Building"].fillna(no_floors_mode)
    return new_data
    

def state_imputations(new_data):
    state_mode = new_data["State"].mode().iloc[0]
    new_data["State"] = new_data["State"].fillna(state_mode)
    state_groupby = new_data[geographical_features + ["State"]].groupby("State").mean()
    return new_data, state_groupby

def geographical_imputations(new_data, geographical_features, state_groupby): 
    for feature in geographical_features: 
        temp_data = new_data.apply(
            lambda x: state_groupby.loc[x["State"], feature] if np.isnan(x[feature]) else x[feature],
            axis=1
        )
        new_data[feature] = temp_data
    return new_data
        
def numerical_imputations(new_data):
    new_data["Base Flood Elevation"] = new_data["Base Flood Elevation"].fillna(0)
    elevation_difference_mean = new_data["Elevation Difference"].mean()
    new_data["Elevation Difference"] = new_data["Elevation Difference"].fillna(elevation_difference_mean)
    policy_count_mode = new_data["Policy Count"].mode().iloc[0]
    new_data["Policy Count"] = new_data["Policy Count"].fillna(policy_count_mode)
    new_data["Lowest Adjacent Grade"] = new_data["Lowest Adjacent Grade"].fillna(0)
    new_data["Lowest Floor Elevation"] = new_data["Lowest Floor Elevation"].fillna(0)
    building_coverage = new_data["Total Building Insurance Coverage"].mean() 
    new_data["Total Building Insurance Coverage"].fillna(building_coverage)
    contents_coverage = new_data["Total Contents Insurance Coverage"].mean()
    new_data["Total Contents Insurance Coverage"].fillna(contents_coverage)
    return new_data
```

```python
new_data = datetime_imputations(new_data)
new_data = boolean_imputations(new_data, boolean_variables)
new_data = categorical_imputations(new_data)
new_data, state_groupby = state_imputations(new_data)
new_data = geographical_imputations(new_data, geographical_features, state_groupby)
new_data = numerical_imputations(new_data)
```
